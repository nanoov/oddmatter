## Preamble

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from importlib import reload
import cmath
```

```python
plt.rcParams['figure.figsize'] = [10, 10]
plt.rcParams['text.usetex'] = True
```

### General definitions

```python
arr_q = np.linspace(-math.pi, math.pi, 200)

def plot_disprel(arromega, a=1):
    q = np.linspace(-math.pi/a, math.pi/a, 200)
    f, ax = plt.subplots()
    ax.set_xlabel(r'$q$', fontsize=15)
    ax.set_ylabel(r'$\omega$', fontsize=15)
    if type(arromega) is list:
        for w in arromega:
            ax.plot(q, w)
    else:
        ax.plot(q, arromega)
```

## Dispertion Relation non-dissipative case

```python
def omega1(q, k, k_a, k_o=0):
    return cmath.sqrt(2 * k * (1 - math.cos(q)) - k_a * math.sin(q) + k_o)
```

```python
def plot_complexfrec1(k=1, k_a=1, k_o=0):
    realomega = np.array([omega1(q, k, k_a, k_o).real for q in arr_q])
    imagomega = np.array([omega1(q, k, k_a, k_o).imag for q in arr_q])
    plot_disprel(arromega=[realomega, imagomega])
```

```python
plot_complexfrec1()
```

```python
widgets.interact(plot_complexfrec1, k=widgets.fixed(1), k_a=(-2,2,0.1), k_o=(0,1,0.1))
```
