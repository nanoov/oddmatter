## Dispertion Relation non-dissipative case

```python
def omega(q, k, k_a, k_o=0):
    return cmath.sqrt(2 * k * (1 - math.cos(q)) - k_a * math.sin(q) + k_o)
```

```python
arr_q = np.linspace(-math.pi, math.pi, 200)

def plot_disprel(arromega):
    f, ax = plt.subplots()
    ax.set_xlabel(r'$q$', fontsize=15)
    ax.set_ylabel(r'$\omega^2$', fontsize=15)
    if type(arromega) is list:
        for w in arromega:
            ax.plot(arr_q, w)
    else:
        ax.plot(arr_q, arromega)
```

```python
k_val = 1
k_a_val = 1
k_o_val = 0

realomega = np.array([omega(q, k_val, k_a_val, k_o_val).real for q in arr_q])
imagomega = np.array([omega(q, k_val, k_a_val, k_o_val).imag for q in arr_q])

plot_disprel(arromega=[realomega, imagomega])
```
